## ASAP

```python
asap_file = r"C:\Users\alvar\Githubs\mentor-eval\datasets\asap-aes\asap_student_responses_and_evaluations.xlsx"
df = pd.read_excel(asap_file)
df.shape
```

```python
df.head()
```

```python
# Create a dictionary to store dataframes by essay_set
essay_sets = {}

# Split the dataframe by essay_set
for set_id in df['essay_set'].unique():
    essay_sets[set_id] = df[df['essay_set'] == set_id]

# For each essay set, print min and max of numeric columns
for set_id, set_df in essay_sets.items():
    print(f"\nEssay Set {set_id}:")
    print("-" * 50)
    
    # Get numeric columns
    numeric_cols = set_df.select_dtypes(include=['int64', 'float64']).columns
    
    # Print min and max for each numeric column
    for col in numeric_cols:
        print(f"{col}:")
        print(f"  Min: {set_df[col].min()}")
        print(f"  Max: {set_df[col].max()}")
    
    # Check all domain scores against their corresponding rater scores
    domain_cols = [col for col in set_df.columns if col.startswith('domain') and col.endswith('_score')]
    
    for domain_col in domain_cols:
        domain_num = domain_col.replace('domain', '').replace('_score', '')
        rater_cols = [col for col in set_df.columns if col.startswith(f'rater') and f'domain{domain_num}' in col]
        
        # Only check if we have both domain score and at least one rater
        if domain_col in set_df.columns and rater_cols:
            # Get non-null domain scores
            valid_scores = set_df[domain_col].notna()
            
            if valid_scores.any():
                if set_id in [1, 7, 8]:
                    # For essay sets 1, 7, and 8, sum all non-null rater scores
                    # Only consider non-null values when summing
                    total_raters = set_df[rater_cols].apply(lambda x: x.dropna().sum(), axis=1)
                    matches = (set_df[domain_col] == total_raters).mean() * 100
                    print(f"\nPercentage where {domain_col} matches sum of rater scores: {matches:.2f}%")
                    
                    # Print an example of non-matching scores
                    non_matches = set_df[domain_col] != total_raters
                    if non_matches.any():
                        example_idx = non_matches[non_matches].index[0]
                        print(f"\nExample of non-matching scores (Row {example_idx}):")
                        print(f"Domain score: {set_df.loc[example_idx, domain_col]}")
                        print(f"Sum of rater scores: {total_raters[example_idx]}")
                        print(f"Individual rater scores: {set_df.loc[example_idx, rater_cols].to_dict()}")
                else:
                    # For other essay sets, calculate average of rater scores
                    # Only consider non-null values when averaging
                    avg_raters = set_df[rater_cols].apply(lambda x: x.dropna().mean(), axis=1)
                    # Consider scores matching if difference is less than 1
                    matches = (abs(set_df[domain_col] - avg_raters) < 1).mean() * 100
                    print(f"\nPercentage where {domain_col} matches average of rater scores (within 1 point): {matches:.2f}%")
                    
                    # Print an example of non-matching scores
                    non_matches = abs(set_df[domain_col] - avg_raters) >= 1
                    if non_matches.any():
                        example_idx = non_matches[non_matches].index[0]
                        print(f"\nExample of non-matching scores (Row {example_idx}):")
                        print(f"Domain score: {set_df.loc[example_idx, domain_col]}")
                        print(f"Average of rater scores: {avg_raters[example_idx]:.2f}")
                        print(f"Difference: {abs(set_df.loc[example_idx, domain_col] - avg_raters[example_idx]):.2f}")
                        print(f"Individual rater scores: {set_df.loc[example_idx, rater_cols].to_dict()}")
```
